```python
import pandas as pd
df1 = pd.read_csv('AMC/AMC_Monthly_stock_price_5y.csv')
df2 = pd.read_csv('CGX/CGX_Monthly_stock_price_5y.csv')
df3 = pd.read_csv('CNK/CNK_Monthly_stock_price_5y.csv')
display(df1,df2,df3)
```

```python

# Function to process each DataFrame
def process_unnamed_columns(df):
    # List of columns in the DataFrame
    existing_columns = df.columns

    # List of columns to drop
    cols_to_drop = [col for col in existing_columns if 'Unnamed' in col]

    # Checking if each column exists before dropping
    for col in cols_to_drop:
        if col in existing_columns:
            df.drop(col, axis=1, inplace=True)

# Apply the function to each DataFrame
process_unnamed_columns(df1)
process_unnamed_columns(df2)
process_unnamed_columns(df3)

# Display the DataFrames after processing
display(df1,df2,df3)
```

```python
import pandas as pd

def process_conversion(df, conversion_factor, columns_to_convert):
    for col in columns_to_convert:
        # Convert to numeric, multiply by conversion factor, and round to the nearest hundredth
        df[col] = (pd.to_numeric(df[col], errors='coerce') * conversion_factor).round(2)
    return df

# Load data from CSV
df2 = pd.read_csv('CGX/CGX_Monthly_stock_price_5y.csv')

# Define the conversion factor (CAD to USD)
conversion_factor = 0.729041  # As per the exchange rate provided

# Specify the columns to convert
columns_to_convert = ['Open', 'High', 'Low', 'Close', 'Adj Close']

# Apply the conversion
df2 = process_conversion(df2, conversion_factor, columns_to_convert)

# Save the modified DataFrame back to CSV
df2.to_csv('CGX/CGX_Monthly_stock_price_5y_converted.csv', index=False)

print(df2.head())
```

```python
import pandas as pd

def process_conversion(df, conversion_factor, columns_to_convert):
    for col in columns_to_convert:
        df[col] = pd.to_numeric(df[col], errors='coerce') / conversion_factor
    return df

# Load data from CSV
df1 = pd.read_csv('AMC/AMC_Monthly_stock_price_5y.csv')

# Define the conversion factor
conversion_factor = 10

# Specify the columns to convert
columns_to_convert = ['Open', 'High', 'Low', 'Close', 'Adj Close']

# Apply the conversion
df1 = process_conversion(df1, conversion_factor, columns_to_convert)

# Save the modified DataFrame back to CSV
df1.to_csv('AMC/AMC_Monthly_stock_price_5y_converted.csv', index=False)

print(df1.head())
```
