```python
import pandas as pd
df1 = pd.read_csv('AMC/AMC_Earnings.csv')
df2 = pd.read_csv('CGX/CGX_Earnings.csv')
df3 = pd.read_csv('CNK/CNK_Earnings.csv')
display(df1,df2,df3)
```

```python
def process_revenue_columns(df):
    cols_divide = [col for col in df.columns if 'Revenue' in col]

    for col in cols_divide:
        # Remove non-numeric characters like commas, 'M' and periods from the columns
        df[col] = df[col].replace({'M': '', ',': '', '\.': ''}, regex=True)

        # Convert the column to numeric, set non-convertible values to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

        # Divide the column values by 1 million if needed
        # If already in millions and M was removed, this division isn't needed
        # Check if any values are larger than a million first
        if df[col].max() > 999.999:
            df[col] = df[col] / 1_000_000

        # Rename the column to indicate 'in millions'
        new_col_name = col + ' (in millions)'
        df.rename(columns={col: new_col_name}, inplace=True)

# Assuming df1, df2, and df3 are already defined DataFrames
process_revenue_columns(df1)
process_revenue_columns(df2)
process_revenue_columns(df3)

# Display the updated DataFrames to verify changes
display(df1)
display(df2)
display(df3)

```

```python
# Define the new standard column names
standard_columns = ['Period', 'EPS', 'EPS Beat/Miss', 'Revenue (in millions)', 'YoY', 'Revenue Beat/Miss2 (in millions)', 'Company']

# Rename the columns for each DataFrame to match the standard
df1.columns = standard_columns
df2.columns = standard_columns
# For df3, it seems 'Earnings History' should be 'Period' to match the other DataFrames
df3.columns = standard_columns
```

```python
# Add the company identifier
df1['Company'] = 'AMC'
df2['Company'] = 'CGX'
df3['Company'] = 'CNK'
```
